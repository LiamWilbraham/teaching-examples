# Matplotlib examples

```python
import matplotlib.pyplot as plt
import numpy as np
```

## scatter graph

```python
x = np.arange(0.0, 50.0, 2.0)
y1 = x ** 1.3 + np.random.rand(*x.shape) * 30.0
y2 = x ** 1.3 + np.random.rand(*x.shape) * 30.0

plt.figure(figsize=(4, 4))

plt.scatter(x, y1, color="g", alpha=0.5, marker='o', label='series1')
plt.scatter(x, y2, color="orange", alpha=0.5, marker='o', label='series2')
plt.xlabel("X (units)")
plt.ylabel("Y (Units)")
plt.legend()

plt.savefig('scatter-plot.png', dpi=300)
plt.show()

# you can also change a bunch of other properties of the plot 
# e.g. number of ticks, tick labels, font size etc but there are 
# so many that it is best just to google
```

## line graph

```python
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0

plt.figure(figsize=(4, 4))

plt.scatter(x, y, color="g", alpha=0.5, marker='o', label='series1')
plt.plot(x, y, color="b")   # <--- add this to get connecting lines 
plt.xlabel("X (units)")
plt.ylabel("Y (Units)")
plt.legend()

plt.savefig('line-plot.png', dpi=300)
plt.show()
```

## histograms (1D and 2D)

```python
# 1D Histogram
x = np.random.randn(100000)

plt.figure(figsize=(4, 4))
plt.hist(x, bins=50)
plt.xlabel("X (units)")
plt.ylabel("Y (Units)")

plt.savefig('1d-hist-plot.png', dpi=300)
plt.show()

# 2D Histogram
x = np.random.randn(100000)
y = np.random.randn(100000) + np.random.randint(500)

plt.figure(figsize=(5, 4))
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.afmhot, label='test')
plt.xlabel("X (units)")
plt.ylabel("Y (Units)")
plt.colorbar()

plt.savefig('2d-hist-plot.png', dpi=300)
plt.show()
```

## multiple plots on a figure

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))

# =================================================================== FIG1
x = np.arange(0.0, 50.0, 2.0)
y1 = x ** 1.3 + np.random.rand(*x.shape) * 30.0
y2 = x ** 1.3 + np.random.rand(*x.shape) * 30.0

ax1.scatter(x, y1, color="g", alpha=0.5, marker='o', label='series1')
ax1.scatter(x, y2, color="orange", alpha=0.5, marker='o', label='series2')
ax1.set_xlabel("X (units)")
ax1.set_ylabel("Y (Units)")
ax1.legend()


# =================================================================== FIG2
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0

ax2.scatter(x, y, color="g", alpha=0.5, marker='o', label='series1')
ax2.plot(x, y, color="b")   # <--- add this to get connecting lines 
ax2.set_xlabel("X (units)")
ax2.set_ylabel("Y (Units)")
ax2.legend()


# =================================================================== FIG3
x = np.random.randn(100000)

ax3.hist(x, bins=50)
ax3.set_xlabel("X (units)")
ax3.set_ylabel("Y (Units)")


# =================================================================== FIG4
x = np.random.randn(100000)
y = np.random.randn(100000) + np.random.randint(500)

ax4.hist2d(x, y, bins=(50, 50), cmap=plt.cm.afmhot, label='test')
ax4.set_xlabel("X (units)")
ax4.set_ylabel("Y (Units)")

plt.savefig('multiple-plots.png', dpi=300)
plt.show()
```
